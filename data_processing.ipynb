```python
import pandas as pd
import numpy as np
import os
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
accidents_data = pd.read_csv("./cleaned_accidents_data.csv")
accidents_data.columns
```

```python
accidents_data.drop(columns="Unnamed: 0",inplace=True)
```

```python
accidents_data
```

```python
accidents_data.dtypes
```

```python
engine = create_engine("sqlite:///accidents.sqlite")
Base = declarative_base()
```

```python

class CleanAccident(Base):
    __tablename__ = 'CleanedAccident'
    
    id = Column(Integer, primary_key=True)
    case_id = Column(Integer)
    date = Column(Integer)
    state = Column(Integer)
    latitude = Column(Float)
    longitude = Column(Float)
    drunk_drivers = Column(Integer)
    fatalities = Column(Integer)
    State = Column(Text)
    

```

```python
Base.metadata.create_all(engine)
```

```python
accidents_dict = accidents_data.to_dict(orient='records')
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
accidents_table = sqlalchemy.Table('CleanedAccident', metadata, autoload=True)
```
